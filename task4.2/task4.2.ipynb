Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Preparing array of independent variable (X) and array of dependent variable(y)

```python
dataset = pd.read_csv('shrink_ray_dataset.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Creating linear regression model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

Required knob location

```python
required_knob_adjustion = (85 - regressor.intercept_)/ regressor.coef_
print(required_knob_adjustion)
```

Plotting data points and linear model

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Shrinkage model')
plt.xlabel('Knob(MW)')
plt.ylabel('Shrinkage(%)')
plt.show()
```
